## Adicionando a coluna "classe_geral" referente à coluna "localizacao"

```python
df['classe_geral'] = None
df.loc[(df['localizacao'] >= 0) & (df['localizacao'] <= 99), 'classe_geral'] = 'Generalidades. Ciência e conhecimento'
df.loc[(df['localizacao'] >= 100) & (df['localizacao'] <= 199), 'classe_geral'] = 'Filosofia e psicologia'
df.loc[(df['localizacao'] >= 200) & (df['localizacao'] <= 299), 'classe_geral'] = 'Religião'
df.loc[(df['localizacao'] >= 300) & (df['localizacao'] <= 399), 'classe_geral'] = 'Ciências sociais'
df.loc[(df['localizacao'] >= 400) & (df['localizacao'] <= 499), 'classe_geral'] = 'Classe vaga. Provisoriamente não ocupada'
df.loc[(df['localizacao'] >= 500) & (df['localizacao'] <= 599), 'classe_geral'] = 'Matemática e ciências naturais'
df.loc[(df['localizacao'] >= 600) & (df['localizacao'] <= 699), 'classe_geral'] = 'Ciências aplicadas'
df.loc[(df['localizacao'] >= 700) & (df['localizacao'] <= 799), 'classe_geral'] = 'Belas artes'
df.loc[(df['localizacao'] >= 800) & (df['localizacao'] <= 899), 'classe_geral'] = 'Linguagem. Língua. Linguística'
df.loc[(df['localizacao'] >= 900) & (df['localizacao'] <= 999), 'classe_geral'] = 'Geografia. Biografia. História'
```
