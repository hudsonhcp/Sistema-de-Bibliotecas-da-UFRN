## Graduação X Pós Graduação

### Total de Empréstimos

```python
df.head()
```

```python
emprestimos_por_tipo = pd.DataFrame({
    'tipo_vinculo_usuario': df['tipo_vinculo_usuario'][(df['tipo_vinculo_usuario'] == 'ALUNO DE GRADUAÇÃO')|(df['tipo_vinculo_usuario'] == 'ALUNO DE PÓS-GRADUAÇÃO')],
    'mes_emprestimo': df['data_emprestimo'].dt.month})
```

```python
emprestimos_por_tipo.head()
```

```python
emprestimo_mensal_grad = emprestimos_por_tipo[['mes_emprestimo']][emprestimos_por_tipo['tipo_vinculo_usuario'] == 'ALUNO DE GRADUAÇÃO'].value_counts().to_frame('total_emprestimos').sort_values(by = ['mes_emprestimo'])
emprestimo_mensal_pos = emprestimos_por_tipo[['mes_emprestimo']][emprestimos_por_tipo['tipo_vinculo_usuario'] == 'ALUNO DE PÓS-GRADUAÇÃO'].value_counts().to_frame('total_emprestimos').sort_values(by = ['mes_emprestimo'])
```

```python
emprestimo_mensal_grad = emprestimo_mensal_grad.rename_axis('mes_emprestimo').reset_index()
emprestimo_mensal_pos = emprestimo_mensal_pos.rename_axis('mes_emprestimo').reset_index()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (16,4))
ax[0].plot(emprestimo_mensal_grad['mes_emprestimo'], emprestimo_mensal_grad['total_emprestimos'], color = 'green')
ax[0].set_title('Empréstimo Mensal dos alunos da Graduação')
ax[0].axhline(emprestimo_mensal_grad['total_emprestimos'].mean(), color = 'orange', label = f"Média/Graduação {emprestimo_mensal_grad['total_emprestimos'].mean():.2f}")
ax[1].plot(emprestimo_mensal_pos['mes_emprestimo'], emprestimo_mensal_pos['total_emprestimos'], color = 'blue',)
ax[1].axhline(emprestimo_mensal_pos['total_emprestimos'].mean(), color = 'orange', label = f"Média/Pós {emprestimo_mensal_pos['total_emprestimos'].mean():.2f}")
ax[1].set_title('Empréstimo Mensal dos alunos de Pós-Graduação')
fig.legend()
fig.show()
```
