## Adicionando a coluna "classe_geral" referente à coluna "localizacao"

### Analisando a quantidade de empréstimos por ano

```python
total_emprestimos = df[['id_exemplar', 'data_emprestimo']].copy()
```

```python
total_emprestimos[['id_exemplar']].value_counts()
```

```python
emprestimos_por_ano = total_emprestimos['data_emprestimo'].dt.year.value_counts().to_frame('total_emprestimos_ano')
```

```python
(emprestimos_por_ano.index.rename(name = 'ano', inplace = True))
```

```python
emprestimos_por_ano = emprestimos_por_ano.sort_values(by='ano')
```

```python
emprestimos_por_ano
```

```python
emprestimos_por_ano.plot()
plt.axhline(emprestimos_por_ano['total_emprestimos_ano'].max(), color = 'green', label = 'Máximo')
plt.axhline(emprestimos_por_ano['total_emprestimos_ano'].min(), color = 'red', label = 'Mínimo')
plt.axhline(emprestimos_por_ano['total_emprestimos_ano'].mean(), color = 'yellow', label = 'Média')
plt.legend()
plt.grid()
plt.title('Empréstimos por Ano')

plt.show()
```

### Analisando a quantidade de empréstimos por mês

```python
emprestimos_por_mes = total_emprestimos['data_emprestimo'].dt.month.value_counts().to_frame('total_emprestimos_mes')
```

```python
emprestimos_por_mes
```

```python
emprestimos_por_mes.index.rename(name = 'mes', inplace = True)
```

```python
emprestimos_por_mes = emprestimos_por_mes.sort_values(by='mes')
```

```python
emprestimos_por_mes
```

```python
emprestimos_por_mes.plot()
plt.axhline(emprestimos_por_mes['total_emprestimos_mes'].max(), color = 'green', label = 'Máximo')
plt.axhline(emprestimos_por_mes['total_emprestimos_mes'].min(), color = 'red', label = 'Mínimo')
plt.legend()
plt.grid()
plt.title('Empréstimos por Mês')

plt.show()
```

### Analisando a quantidade de empréstimos por hora

```python
emprestimos_por_hora = total_emprestimos['data_emprestimo'].dt.hour.value_counts().to_frame('total_emprestimos_hora')
```

```python
emprestimos_por_hora
```

```python
emprestimos_por_hora.index.rename(name = 'hora', inplace = True)
```

```python
emprestimos_por_hora = emprestimos_por_hora.sort_values(by='hora', ascending = True)
```

```python
emprestimos_por_hora
```

```python
emprestimos_por_hora.plot(kind = 'bar')
plt.axhline(emprestimos_por_hora['total_emprestimos_hora'].max(), color = 'green', label = 'Máximo')
plt.axhline(emprestimos_por_hora['total_emprestimos_hora'].min(), color = 'red', label = 'Mínimo')
plt.legend()
plt.grid()
plt.title('Empréstimos por Hora')

plt.show()
```
